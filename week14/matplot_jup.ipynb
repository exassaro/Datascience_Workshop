# Matplotlib basic plotting

Line Plot

```python
import matplotlib.pyplot as plt
X=[0,1,2,3,4,5,6]
Y=[0,1,4,9,16,25,36]

plt.plot(X,Y,label="x=x^2",color="purple",marker='o')
plt.xlabel('Integer')
plt.ylabel('Square')
plt.title('Line Plot Practice')
plt.grid("True")
plt.legend()
plt.show()
```

Bar chart

```python
categories=['Math','Sci','Eng','Pe']
scores=[90,80,67,45]

plt.bar(categories,scores,color="blue")
plt.title("Student Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.show()
```

Subplots challenge

```python
fig,axs =plt.subplots(1,2,figsize=(10,4))
axs[0].bar(['A','B','C'],[3,7,5])
axs[0].set_title('Bar Chart')

axs[1].plot([1,2,3],[1,4,9])
axs[1].set_title('Line Chart')

plt.tight_layout()
plt.show()
```

Pie Chart time

```python
sizes=[40,30,20,10]
labels=['Python','Java','C++','other']

plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90)
plt.title('Language Popularity')
plt.axis('equal')
plt.show()
```

Scatter Plot 

```python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
plt.scatter(x,y,color='red',marker='x')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips =sns.load_dataset("tips")
sns.scatterplot(x="total_bill",y="tip",data=tips) # Plots each data point as a dot
plt.show()
```

sns.lineplot()

```python
df=sns.load_dataset('tips')
sns.scatterplot(x="total_bill",y="tip",data=tips) # Plots each data point as a dot
plt.title('Line plot of Total Bill vs Tip')
plt.show()
```

sns.barplot()

```python
df=sns.load_dataset('tips')
sns.barplot(x='day',y='total_bill',data=df)
plt.title('Average Total Bill per day') 
plt.show()
```

sns.countplot()

```python
df=sns.load_dataset('tips')
sns.countplot(x='day',data=df)
plt.title('Count of records per day')
plt.show()
```

sns.histplot()

```python
df=sns.load_dataset('tips')
sns.histplot(data=df,x='total_bill')
plt.title('Distribution of total Bill Amount')
plt.show()
```

sns.kdeplot()

```python
df=sns.load_dataset('tips')
sns.kdeplot(data=df,x='total_bill')
plt.title('KDE of total Bill')
plt.show()
```

```python
df=sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill',data=df)
plt.title('Total bill distribution by day')
plt.show()
```

sns.heatmap()

```python
df = sns.load_dataset('tips')
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
