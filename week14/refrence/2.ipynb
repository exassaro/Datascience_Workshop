# data visualization with matplotlib and seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load the three sample datasets
iris=sns.load_dataset('iris')
titanic=sns.load_dataset('titanic')
tips=sns.load_dataset('tips')
iris.head()
```

### matplotlib

```python
# 1 Line plot

plt.figure(figsize=(8,5))
plt.plot(iris['sepal_length'],iris['sepal_width'],marker='o',linestyle='-',color='blue')
plt.title('Line plot: sepal length vs sepal width')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.grid(True)
plt.show()
```

```python
# 2 Bar plot

plt.figure(figsize=(8,5))
plt.bar(tips['day'],tips['total_bill'],color='skyblue')
plt.title('Bar plot: Total bill by day')
plt.xlabel('day')
plt.ylabel('total bill')
plt.grid(True)
plt.show()
```

```python
# 3 Histogram
plt.figure(figsize=(8,5))
plt.hist(titanic['age'].dropna(),bins=30,color='green',edgecolor='black') 
plt.title('Histogram:Age distribution on titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4 Scatter plot
plt.figure(figsize=(8,5))
plt.scatter(iris['petal_length'],iris['petal_width'],color='red')
plt.title('scatter plot : petal length vs petal width')
plt.xlabel('petal length')
plt.ylabel('petal width')
# plt.grid(True)
plt.show()
```

```python
# 5 pie chart

plt.figure(figsize=(6,6))
survived_counts=titanic['survived'].value_counts()
plt.pie(survived_counts,labels=['Not survived','Survived'],autopct='%1.1f%%',colors=['hotpink','lightgreen'])
plt.title('pie chart : titanic survival rate')
plt.show()
```

### seaborn common plotting techniques 

```python
# 1 scatter plot

plt.figure(figsize=(8,5))
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
plt.title('scatter plot:sepal dimensions by species')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
# 2 heatmap
plt.figure(figsize=(8,6))
correlation_matrix=iris.corr(numeric_only=True)
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('heatmap:correlation matrix (iris dataset)')
plt.show()
```

```python
# 3 box plot
plt.figure(figsize=(8,5))
sns.boxplot(x='day',y='total_bill',data=tips,palette='Set2')
plt.title('box plot: total bill by day')
plt.xlabel('day')
plt.ylabel('total bill')
plt.show()
```

```python
# 4 violin plot

plt.figure(figsize=(8,5))
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips,split=True,palette='Set3')
plt.title('violin plot: total bill by day and gender')
plt.xlabel('day')
plt.ylabel('total bill')
plt.show()
```

```python
# 5 pair plot

sns.pairplot(iris,hue='species')
plt.suptitle('pair plot: iris dataset',y=1.02)
plt.show()
```
