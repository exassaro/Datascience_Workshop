```python
import pandas as pd
import seaborn as sns
df=sns.load_dataset('titanic')
print(df)
```

# indexing and selecting data

### loc[] (label base indexing)  (accessing rows)

```python
# selects row at index 1
print(df.loc[1])
```

```python
# selects column 'age'
print(df.loc[:,'age'])
```

```python
# selects multiple rows and columns
print(df.loc[0:1,['age','sex']])
```

### using iloc[] (integere based indexing)

```python
# selects row at position 1
print(df.iloc[1])
```

```python
# selects first column
print(df.iloc[:,0])
```

```python
# selects subsets of rows and columns
print(df.iloc[0:2,0:2])
```

### iat[] (fast access for single value-integer-based)

```python
# fetches the value in row 1,column 1
print(df.iat[1,1])
```

# data types in pandas

### checking data types

```python
print(df.dtypes)
```

### changing data type

```python

df['age']=df['age'].astype(float)
print(df)
```

### accessing columns

```python
# access a single column
df['deck'] #returns a series

print(df.columns)
# access multiple columns
print(df[['pclass','age']]) #returns a dataframe
```

# slicing data (subset of data)

### row slicing

```python
print(df[0:2])
# select rows with index 0 and 1
```

### column slicing

```python
print(df.loc[:,'age':'alive'])
```

### row and column slicing

```python
print(df.iloc[0:2,0:2])
```

# filtering data (applying conditions)

### filtering rows based on condition

```python
df[df['age']>25]
```

### filtering using multiple conditions

```python
df[(df['age']>25) & (df['who']=='man')]
```

```python
df[df['age'].isin([25,35])]
```

```python
df[df['sex'].str.contains('f',case=False)]
```

```python
df[df['sex'].str.contains('F',case=True)]
# case=true make checking only for F in sex not f, case sensitive=true
```

# checking for missing values

```python
print(df.isna())
# returns a dataframe with true where NaN exists

print(df.isnull().sum())
# counts missing values per column
```

# filling missing data

```python
df=df.astype(str)
df_filled=df.fillna('unknown')
print(df_filled)
```

# renaming columns

```python
df=df.rename(columns={'who':'category'})
print(df)
```
