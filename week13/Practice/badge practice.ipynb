```python
#decorator example

import time

def time_it(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"total time for {func.__name__} is {start-end}")
        return result
    return wrapper

@time_it
def square(num1):
    sqr = num1*num1
    return sqr

@time_it
def cube(num1):
    cub = num1*num1*num1
    return cub



        
print(square(5))
print(cube(2))
        
        
        
        
```

```python
#generator function
def gen(max):
    count=1
    while count<=max:
        yield count
        count +=1
ctr = gen(10)#this function returns a iterator from 1 to 10
for i in ctr:
    print(i)
```

```python
#generator function for multiples of 5 under 100
def gen(max):
    result = 5
    while result<=max:
        yield result
        result+=5
five = gen(100)
for i in five :
    print(i)    
```
