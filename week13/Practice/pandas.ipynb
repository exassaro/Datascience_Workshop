# handling missing data

#### check for missing values

```python
df.isnull().sum()
# count missing values in each column
```

#### fill missing values

```python
df.fillna('unknown',inplace=True)
# replace the NaN with 'unknown'
```

#### remove rows with missing values

```python
df.dropna(inplace=True)
# remove rows with NaN values
```

# group and aggregate data

```python
df.groupby('post_day')['likes'].mean()
# average likes per post_day
df['likes'].max()
# find maximum likes
```

# sorting data

```python
df.sort_values('post_hour',ascending=False)
```

# merging and combining data

```python
import seaborn as sns
df2=sns.load_dataset('titanic')
print(df2)
```

#### merge two dataframes

```python
print(df.columns)
df['sex']='male'
print(df2.columns)
```

```python
df.merge(df2,on='sex',how='inner')
# inner join on 'salary'
# only the matching data's will be present

df.merge(df2,on='sex',how='outer')
# merge all the datas and replace NaN with empty cells in the dataset
```

#### concatinate two dataframes

```python
pd.concat([df,df2])
# append dataframes
```
